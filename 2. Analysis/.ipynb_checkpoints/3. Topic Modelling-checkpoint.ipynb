# Topic Modeling
* Each notes is represented as a distribution over topics.
* Each topic is represented as a distribution over words.

```python
import pandas as pd
import pickle
import gensim as gensim
from gensim import corpora
from gensim import matutils, models
import scipy.sparse
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Topic Modeling - Attempt #1 (All Texts)

```python
import pandas as pd
import pickle

data = pd.read_pickle('dtm_stop.pkl')
data
```

```python
# One of the required inputs is a term-document matrix
tdm = data.transpose()
tdm.head()
```

```python
# put the term-document matrix into a new gensim format, from df --> sparse matrix --> gensim corpus
from gensim import matutils, models
import scipy.sparse


sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
# Gensim also requires dictionary of the all terms and their respective location in the term-document matrix
cv = pickle.load(open("cv_stop.pkl", "rb"))
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
# the corpus (term-document matrix) and id2word (dictionary of location: term),
# specify two other parameters as well - the number of topics and the number of passes

# 3 topic
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=50)
lda.save('Ida3.gensim')

lda.print_topics()
```

```python
# 4 topic
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=4, passes=50)
lda.save('Ida4.gensim')

lda.print_topics()
```

```python
# 5 topic
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=5, passes=50)
lda.save('Ida5.gensim')

lda.print_topics()
```

# Topic Modeling - Attempt #2 (Nouns Only)
* Look only at terms that are from part of speech (only nouns, only adjectives, etc.)
* UPenn tag set: https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html.

```python
# create a function to pull out nouns from a string of text
from nltk import word_tokenize, pos_tag

def nouns(text):
    '''Given a string of text, tokenize the text and pull out only the nouns.'''
    is_noun = lambda pos: pos[:2] == 'NN'
    tokenized = word_tokenize(text)
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)] 
    return ' '.join(all_nouns)
```

```python
# Read in the cleaned data, before the CountVectorizer step
data_clean = pd.read_pickle('data_clean.pkl')
data_clean
```

```python
# Apply the nouns function to the transcripts to filter nouns

#data_clean = data_clean['text'].apply(str)
#data_nouns = pd.DataFrame(data_clean.apply(nouns))
data_nouns = data_clean.apply(nouns)
data_nouns
```

```python
# Create a new document-term matrix using only nouns
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Re-add the additional stop words since we are recreating the document-term matrix
add_stop_words = ['i', 'im', 'thing','dont','year','time','day','way',
                 'que','eu','edge','fuck']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Recreate a document-term matrix with only nouns
cvn = CountVectorizer(stop_words=stop_words,max_df=.8)
data_cvn = cvn.fit_transform(data_nouns)
data_dtmn = pd.DataFrame(data_cvn.toarray(), columns=cvn.get_feature_names())
data_dtmn.index = data_nouns.index
data_dtmn
```

```python
# Create the gensim corpus
corpusn = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmn.transpose()))

# Create the vocabulary dictionary
id2wordn = dict((v, k) for k, v in cvn.vocabulary_.items())
```

```python
# 3 topics
ldan = models.LdaModel(corpus=corpusn, num_topics=3, id2word=id2wordn, passes=100)
ldan.save('Idan3.gensim')

ldan.print_topics()
```

```python
# 4 topics 
ldan = models.LdaModel(corpus=corpusn, num_topics=4, id2word=id2wordn, passes=100)
ldan.save('Idan4.gensim')

ldan.print_topics()
```

```python
# 5 topics
ldan = models.LdaModel(corpus=corpusn, num_topics=5, id2word=id2wordn, passes=100)
ldan.save('Idan5.gensim')

ldan.print_topics()
```

# Topic Modeling - Attempt #3 (Nouns and Adjectives)

```python
# create a function to pull out nouns from a string of text
def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
# Apply the nouns and adjectives function to the transcripts to filter nouns and adjectives

data_nouns_adj = data_clean.apply(nouns_adj)
data_nouns_adj
```

```python
# Create a new document-term matrix using nouns and adjectives
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Re-add the additional stop words since we are recreating the document-term matrix
add_stop_words = ['i', 'im', 'thing','dont','year','time','day','way',
                 'que','eu','edge','ive','fuck']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)


# Create a new document-term matrix using only nouns and adjectives, also remove common words with max_df
cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names())
data_dtmna.index = data_nouns_adj.index
data_dtmna
```

```python
# Create the gensim corpus
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

# Create the vocabulary dictionary
id2wordna = dict((v, k) for k, v in cvna.vocabulary_.items())
```

```python
# 3 topics
ldana = models.LdaModel(corpus=corpusna, num_topics=3, id2word=id2wordna, passes=100)
ldana.save('Idana3.gensim')

ldana.print_topics()
```

```python
# 4 topics
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=100)
ldana.save('Idana4.gensim')

ldana.print_topics()
```

```python
# 5 topics
ldana = models.LdaModel(corpus=corpusna, num_topics=5, id2word=id2wordna, passes=100)
ldana.save('Idana5.gensim')

ldana.print_topics()
```

# Topic Modeling - Attempt #4 (Nouns and Verbs)

```python
# create a function to pull out nouns and verbs from a string of text
def nouns_vbs(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and verbs.'''
    is_noun_vbs = lambda pos: pos[:2] == 'NN' or pos[:2] == 'VB'
    tokenized = word_tokenize(text)
    nouns_vbs = [word for (word, pos) in pos_tag(tokenized) if is_noun_vbs(pos)] 
    return ' '.join(nouns_vbs)
```

```python
# Apply the nouns and verbs function to the transcripts to filter on nouns and verbs

data_nouns_vbs = data_clean.apply(nouns_vbs)
data_nouns_vbs
```

```python
# Create a new document-term matrix using only nouns and verbs
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Re-add the additional stop words since we are recreating the document-term matrix
add_stop_words = ['i', 'im', 'thing','dont','year','time','day','way',
                 'que','eu','edge','ive','fuck','make','want','know','try']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)


# Create a new document-term matrix using only nouns and verbs, also remove common words with max_df
cvnai = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvnai = cvnai.fit_transform(data_nouns_vbs)
data_dtmnai = pd.DataFrame(data_cvnai.toarray(), columns=cvnai.get_feature_names())
data_dtmnai.index = data_nouns_vbs.index
data_dtmnai
```

```python
# Create the gensim corpus
corpusnai = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmnai.transpose()))

# Create the vocabulary dictionary
id2wordna = dict((v, k) for k, v in cvnai.vocabulary_.items())
```

```python
# 3 topics
ldanai = models.LdaModel(corpus=corpusnai, num_topics=3, id2word=id2wordna, passes=100)
ldanai.save('Idanai3.gensim')

ldanai.print_topics()
```

```python
# 4 topics
ldanai = models.LdaModel(corpus=corpusnai, num_topics=4, id2word=id2wordna, passes=100)
ldanai.save('Idanai4.gensim')

ldanai.print_topics()
```

```python
# 5 topics
ldanai = models.LdaModel(corpus=corpusnai, num_topics=5, id2word=id2wordna, passes=100)
ldanai.save('Idanai5.gensim')

ldanai.print_topics()
```

# Identify Topics in Each Suicidal Notes

```python
# Choose the best and make sense model

Lda_model = models.ldamodel.LdaModel.load('Idan4.gensim')
Lda_model.show_topics()
```

```python
# Let's take a look at which topics each transcript contains
corpus_transformed = Lda_model[corpusn]
post_lists = data_dtmn.index
# zipped = zip([a for [(a,b)] in corpus_transformed])
# list(zip([a for [(a,b)] in corpus_transformed], term_lists))
topics = []
for ele in corpus_transformed:
    topic = max(ele, key = lambda x:x[1])[0]
    topics.append(topic)
# print(topics)
topic_df = pd.DataFrame({'text':data_nouns, 'topic': topics})
topic_df
```

```python
# percentage of the topic
topic_counts =topic_df['topic'].value_counts(normalize=True)
topic_counts
```

```python
# piechart
colors = ("deeppink","turquoise","dodgerblue","gold","mediumaquamarine")
plt.pie(topic_counts, labels = topic_counts.index, autopct='%1.0f%%',colors=colors) #,textprops = dict(color ="grey"))
fig = plt.gcf()
fig.set_size_inches(7,7)
```

# Insight : Emotions in Topics
* Topic 2 contain more emotions in sadness and fear, less in anger 

```python
# save file
topic_post = pd.DataFrame(topic_df[['text','topic']])
topic_post.to_pickle('topic.pkl')
```

```python
# Read file
# emotion by counts
emotion = pd.read_pickle('emotion.pkl')

# emotions by scores
emotion_score = pd.read_pickle('emotion_score.pkl')

# topic file
topic = pd.read_pickle('topic.pkl')
```

## Emotions by domain

```python
topic_emo = pd.concat([topic,emotion],axis=1)
topic_emo.rename(columns={"text":"Text","topic":"Topic","domain_emotion_by_count": "Emotion"}, inplace=True)
topic_emo
```

```python
topic_emo = topic_emo[['Topic','Emotion']]
summary = topic_emo.groupby(['Topic','Emotion']).size().unstack()
summary
```

```python
summary1 = topic_emo.groupby('Topic')['Emotion'].value_counts(normalize=True).unstack()
summary1
```

```python
summary1 = summary1[['Sadness',"Fear","Anger","Surprise","Disgust"]]
summary1
```

```python
#stack barplot
colors = ("powderblue","thistle","lightcoral","khaki","mediumaquamarine")

ax = summary1.apply(lambda x: x*100/sum(x), axis=1).plot(kind="bar", stacked=True, figsize=(10,5), width=0.8, color=colors)

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.text(x+width/2, 
            y+height/2, 
            '{:.0f} %'.format(height), 
            horizontalalignment='left', 
            verticalalignment='center')
    
plt.title("Emotions in Topics")
plt.xlabel("Topic")
plt.ylabel("Percentage (%)")
```
